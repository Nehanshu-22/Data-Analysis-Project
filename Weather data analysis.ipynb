```python
#Weather data analysis
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt

file_data=pd.read_csv("weatherHistory.csv")
file_data
```

```python
file_data[["Date-Time","Time_Zone"]]=file_data["Formatted Date"].str.split("+",expand=True)
df1=file_data.drop(columns="Formatted Date")
df1.head()

# Changing the order of the columns
columns_order=["Date-Time","Time_Zone","Summary","Precip Type","Temperature (C)","Apparent Temperature (C)",
                "Humidity","Wind Speed (km/h)","Wind Bearing (degrees)","Visibility (km)","CLoud Cover",
                "Pressure (millibars)", "Daily Summary"]
df2=file_data.reindex(columns=columns_order)
df3=df2.drop(columns="Time_Zone")
df3.head()
```

```python

df3["Date-Time"]=pd.to_datetime(df3["Date-Time"])
df3["Year"]=pd.DatetimeIndex(df3["Date-Time"]).year
df3["Month"]=df3["Date-Time"].dt.month_name()
df3["day"]=df3["Date-Time"].dt.day
df3.head()
```

```python
# Average Temperature over the last few years

df3['Temperature (C)'].describe()
```

```python
Avg_Temp=pd.DataFrame(df3.groupby("Year")["Temperature (C)"].mean())
Avg_Temp
```

```python
fig,ax=plt.subplots(figsize=(10,10))
sn.lineplot(x=Avg_Temp.index,y=Avg_Temp["Temperature (C)"])
plt.title("Average Annual Temperature")
```

```python
# Average monthly temperature over the years
Avg_Temp_month=pd.DataFrame(df3.groupby("Month")["Temperature (C)"].mean())
month=["January","February","March","April","May","June","July","August","September","October","November","December"]
Avg_Temp_monthly=Avg_Temp_month.reindex(index=month)
Avg_Temp_monthly
```

```python
fig,ax=plt.subplots(figsize=(10,10))
sn.lineplot(x=Avg_Temp_monthly.index,y=Avg_Temp_monthly["Temperature (C)"])
plt.title("Average Monthly Temperature over the years")
```

```python
# Monthly Visibility
Avg_Visib_month=pd.DataFrame(df3.groupby("Month")["Visibility (km)"].mean())
month=["January","February","March","April","May","June","July","August","September","October","November","December"]
Avg_Visib_monthly=Avg_Visib_month.reindex(index=month)
Avg_Visib_monthly
```

```python
fig,ax=plt.subplots(figsize=(10,10))
sn.barplot(x=Avg_Visib_monthly.index,y=Avg_Visib_monthly["Visibility (km)"])
plt.title("Average Monthly Visibility over the years")
```

```python
# Annual Average Wind speed over the years
Avg_Wind_Speed=pd.DataFrame(df3.groupby("Year")["Wind Speed (km/h)"].mean())
Avg_Wind_Speed
```

```python
fig,ax=plt.subplots(figsize=(10,10))
sn.lineplot(data=df3, x ="Year", y="Wind Speed (km/h)")
plt.title("Average Annual Wind Speed over the years")
```

```python
# Relationship between Wind Speed and Humidity
fig = plt.figure(figsize=(10,10))
sn.jointplot(x=df3['Wind Speed (km/h)'],y=df3['Humidity'],kind='kde');
plt.savefig("Wind Speed vs Humidty Density Plot")
```

```python
# Yearly fluctuations in Humidity
fig = plt.figure(figsize=(10,10))
sn.histplot(data = df3, y='Humidity',x='Year',kde = True,hue="Year",color="")
```

```python
# Daily variations in Air Pressure
fig,ax=plt.subplots(figsize=(10,10))
sn.barplot(data=df3, x ="day", y= "Pressure (millibars)")
plt.title("Daily Variations in Air Pressure")
```

```python
# Relationship between Wind Speed and Pressure
fig = plt.figure(figsize=(10,10))
sn.jointplot(x=df3['Wind Speed (km/h)'],y=df3['Pressure (millibars)'],kind='kde');
plt.savefig("Wind Speed vs Pressure Density Plot")
```

```python
# Correlation between variables
df4=df3.drop(["CLoud Cover","Year","day","Month"],axis=1)
d_corr=df4.corr()
d_corr
```

```python
# Correlation heat map
fig,ax=plt.subplots(figsize=(10,10))
sn.heatmap(d_corr,annot=True,cmap='magma_r',linewidths=0.2)
plt.title("correlations heat map")
```
